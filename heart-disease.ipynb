```python
import pandas as pd

from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

import pickle
```

```python
#reading the data
df = pd.read_csv('heart.csv')
df
```

```python
df.describe()
```

```python
#checking for null values
df.isnull().any()
```

```python
df.dtypes
```

```python
#transforming columns of dtype object to numerical
mapping_dict = {}
for cat in [col for col in df.columns if df[col].dtype == 'O']:
    le = preprocessing.LabelEncoder()
    df[cat] = le.fit_transform(df[cat])
    mapping_dict["{0}".format(cat)] = dict(zip(le.classes_, range(len(le.classes_))))
mapping_dict
```

```python
#testing and training splits
X = df.iloc[:, range(0,11)]
y = df.iloc[:, 11]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
#training the model
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
print(classification_report(y_test, rf.predict(X_test)))
```

```python
#exporting the model
pickle.dump(rf, open(('model.pkl'), 'wb'))
```
